```python
j = data.to_json()
```

```python
j = json.loads(j)
```

```python
cleaned = {}
for k in j.keys():
    cleaned[k] = []
    for i in range(3000):
        cleaned[k].append(int(j[k][str(i)]))
```

```python
with open('data.json', 'w') as f:
    f.write(json.dumps(cleaned))
```
